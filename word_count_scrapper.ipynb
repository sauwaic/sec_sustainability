```python
import pandas as pd

sp_report_urls = pd.read_csv('save2.csv')
sp_report_urls.tail()
```

```python
sp_reports_urls = sp_report_urls.drop('Unnamed: 0', axis=1)
sp_reports_urls.info()
```

```python
import requests
import re
from collections import Counter
import time

word_list = [
    'sustainability'
    , 'sustainable'
    , 'renewable'
    , 'climate change'
    , 'green energy'
    , 'carbon neutral'
    , 'greening'
    , 'clean energy'
    , 'esg'
    , 'greenhouse gas'
    , 'recycle'
    , 'circular economy'
    , 'zero emissions'
    , 'environmentally friendly'
]

def count_words(url, word_list):
    counter = Counter()
    text = requests.get(url).text.lower()
    for phrase in word_list:
        counter[phrase] += len(re.findall(r'\b%s\b' % phrase, text))
    return counter

counter_df = pd.DataFrame()
start = time.process_time()

x = sp_report_urls

for i, url in enumerate(x['URL']):
    counter = count_words(url, word_list)
    df = pd.DataFrame(counter, index=[0])
    df['URL'] = url
    counter_df = counter_df.append(df)
    print("%d complete - %.2f secs" %(i, time.process_time() - start))

counter_df = counter_df.reset_index(drop=True)
counter_df.to_csv('word_count.csv')
counter_df.tail()
```
