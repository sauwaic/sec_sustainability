```python
import pandas as pd

sec_report_list = pd.read_csv('sp_reports_links.csv')
word_count = pd.read_csv('word_count.csv')
```

```python
sec_report_list = sec_report_list.drop('Unnamed: 0', axis = 1)
sec_report_list.head()
```

```python
sec_report_list.info()
```

```python
word_count = word_count.drop('Unnamed: 0', axis = 1)
word_count.head()
```

```python
word_count.info()
```

```python
df_combined = pd.merge(sec_report_list, word_count, on='URL')
```

```python
df_combined.head()
```

```python
df_combined.info()
```

```python
df_combined['Date Filed'] = pd.to_datetime(df_combined['Date Filed'], utc=False)
```

```python
df_combined.info()
```

```python
col_list = [
    'sustainability'
    , 'sustainable'
    , 'renewable'
    , 'climate change'
    , 'green energy'
    , 'carbon neutral'
    , 'greening'
    , 'clean energy'
    , 'esg'
    , 'greenhouse gas'
    , 'recycle'
    , 'circular economy'
    , 'zero emissions'
    , 'environmentally friendly'
]

df_combined['sum'] = df_combined[col_list].sum(axis=1)
```

```python
df_combined.head()
```

```python
total_word_count = df_combined.groupby(df_combined['Date Filed'].dt.to_period("Y"))['sum'].sum()
total_word_count
```

```python
ax = total_word_count.plot.bar(y='sum', rot=0)
ax
```

# Top 10 Companies

```python
top_companies = df_combined.groupby(['CIK', 'Company Name'])['sum'].sum().nlargest(10)
top_companies
```

```python
df_combined['company_name'] = df_combined.groupby('CIK')['Company Name'].transform('last')
df_combined.head()
```

```python
import numpy as np

company_pivot = pd.pivot_table(df_combined,
                   index = ['CIK', 'company_name'], 
                   columns = df_combined['Date Filed'].dt.year,
                   values='sum',aggfunc=np.sum)
company_pivot.head()
```

```python
company_pivot['sum'] = company_pivot.sum(axis=1)
```

```python
company_pivot.sort_values(by = ['sum'], ascending = 0).head(10)
```
